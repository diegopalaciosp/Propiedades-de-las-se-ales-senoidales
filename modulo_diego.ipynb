```python
from math import sin, cos, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amplitude = 1.
frequency = 1.
phase = 1.
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitude, frequency, phase, time) for time in x]
plt.style.use('dark_background')
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
import scipy.signal as sc

def sine_wave(A: float, f: float, phi: float, t: float,wl: float) -> float:
    return A*sin(2*pi*wl*f*t + phi)

def ind_max_fun(x):
    i = 0
    maxs = []
    max = 0
    for i in range(len(x)):
        if(x[i] >= max):
            maxs.append(i)
            max = x[i]
    return maxs

amplitude = 1.
frequency = 1.
phase = 1.
wave_length = .5
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitude, frequency, phase, time,wave_length) for time in x]
# wave length velocity/frequency
local_max = ind_max_fun(y)
plt.style.use('dark_background')
plt.plot(x,y)
plt.plot(x[local_max[len(local_max)-2]],y[local_max[len(local_max)-2]],'ro-')
plt.plot(x[local_max[len(local_max)-3]],y[local_max[len(local_max)-3]],'ro-')
plt.ylabel('Amplitud')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.show()
```
