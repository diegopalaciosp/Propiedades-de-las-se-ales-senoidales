# Propiedades de las señales senoidales
## Introducción 
### ¿Qué es una señal senoidal?
También conocida como la sinusoide. Es una curva continua definida en términos de la función trigonométrica seno. 
Las señales senoidales son importantes en las fuentes de información porque fenómenos naturales producen señales que corresponden a señales senoidales como una función del tiempo.
#### Modelo matemático
La señal senoidal es un modelo matemático que se representa mediante la fórmula: Asin(2ft+o)
##### Parámetros
A
Es la amplitud de la señal. La diferencia entre la altura máxima y mínima de la señal. 
2
Son los 2pi radianes que requiere la función para ser una señal senoidal completa. 
f 
Es la frecuencia de la señal. El número de oscilaciones por unidad de tiempo. 
t
Es el tiempo de la señal. 
o
Es la fase de la señal. Que tanto el comienzo de la señal es movida desde un tiempo de referencia.
## Uso del notebook (Modificación de parámetros)
Para utilizar el notebook, se tendrán que modificar las variables indicadas en los markdowns

```python
from math import sin, cos, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
# if using a jupyter notebook
%matplotlib inline

def sine_wave(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amplitude = 1.
frequency = 1.
phase = 1.
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitude, frequency, phase, time) for time in x]
plt.style.use('dark_background')
plt.plot(x,y)
plt.grid(True)
plt.show()
```

## Señal periódica (Explicación)
### ¿Por qué la señal senoidal es periódica?
La señal senoidal es una señal periódica, es decir, que exhibe repetición. Recordemos que las funciones trigonométricas pueden ser representadas en un círculo unitario, de esta manera se aprecia la infinidad de las series que pueden expresar.
### Uso código (Modificación de parámetros)
Cambiese los parámetros debajo de la función "sinu" para modificar la señal

```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

%matplotlib inline

#Graficar el concepto periodo de una señal senoidal
def sinu(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amp = 1
freq = 3
phase = 1
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sinu(amp, freq, phase, time) for time in x]

plt.annotate('Periodo = ', xy=(2.2, 1), xytext=(2.5, 1.5),arrowprops=dict(arrowstyle='->'),)

plt.plot(x,y)
plt.text(3.5, 1.5 , freq)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

## Frecuencia

Es la cantidad de ciclos que ocurren en 1 segundo y se encuentra vinculada al argumento de la función seno: ωt. La frecuencia se denota como f y se mide en ciclos por segundo o Hertz (Hz) en el sistema Internacional.

La frecuencia es la cantidad inversa del período, por lo tanto:

 f = 1/T

## Amplitud

De acuerdo a la expresión general de la onda senoidal v (t) = vm sen (ωt+φ), vm es el valor máximo de la función, que ocurre cuando sen (ωt+φ)= 1 (recordando que el mayor valor que admite tanto la función seno como la función coseno es 1). Este valor máximo es justamente la amplitud de la onda, también conocida como amplitud pico.

En la onda senoidal la amplitud es constante, pero en otros tipos de onda la amplitud puede variar.

## Fase

La fase corresponde al desplazamiento horizontal experimentado por la onda, respecto a un tiempo tomado como referencia.

Dos ondas senoidales están en fase cuando su frecuencia y su fase son las mismas. Si la fase difiere, entonces están en desfase.

Si la frecuencia de las ondas es diferente, estarán en fase cuando la fase ωt+φ sea la misma en ambas ondas en determinados instantes.

### Uso código (Modificación de parámetros)
Cambiese la variable "phase" para notar el efecto de la fase en la onda

## Longitud de onda

La longitud de onda λ se define como la distancia que recorre una onda electromagnética en un tiempo igual a un período. Matemáticamente, esto se traduce en :

λ = c / f m

Dónde:

λ = Longitud de onda.
c = La velocidad de propagación.
f = La frecuencia

Mientras la frecuencia de una señal es independiente del medio, la longitud de onda depende de la frecuencia y del medio.

### Uso código (Modificación de parámetros)
Cambiese la variable "wave_length" para notar el efecto de la longitud de la onda en la señal


```python
import scipy.signal as sc

def sine_wave(A: float, f: float, phi: float, t: float,wl: float) -> float:
    return A*sin(2*pi*wl*f*t + phi)

def ind_max_fun(x):
    i = 0
    maxs = []
    max = 0
    for i in range(len(x)):
        if(x[i] >= max):
            maxs.append(i)
            max = x[i]
    return maxs

amplitude = 1.
frequency = 1.
phase = 1.
wave_length = .5
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitude, frequency, phase, time,wave_length) for time in x]
# wave length velocity/frequency
local_max = ind_max_fun(y)
plt.style.use('dark_background')
plt.plot(x,y)
plt.plot(x[local_max[len(local_max)-2]],y[local_max[len(local_max)-2]],'ro-')
plt.plot(x[local_max[len(local_max)-3]],y[local_max[len(local_max)-3]],'ro-')
plt.ylabel('Amplitud')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.show()
```

## Señal compuesta

Hasta ahora, hemos hablado de las señales periódicas simples, pero existen ondas útiles que no cambian de forma en una única curva entre amplitud máxima y mínima. A estas señales se les conoce como compuestas.

Una señal compuesta está conformada por varias ondas seno simples. Esto significa que una señal compuesta puede ser tanto periódica como aperiódica. Si es periódica se puede descomponer en una serie de ondas seno con frecuencias discretas, frecuencias con valores enteros. En cambio si es aperiódica se puede descomponer en una combinación o en un número infinito de ondas simples con frecuencias continuas, frecuencias con valores reales.

### Uso código (Modificación de parámetros)
Cambiese las variables después de los comentarios "first function" y "second function" para notar otra composición de onda


```python
import scipy.signal as sc

def sine_wave(A: float, f: float, phi: float, t: float,wl: float) -> float:
    return A*sin(2*pi*wl*f*t + phi)

#Parameters first function
amplitude = 1.
frequency = 1.
phase = 0.
wave_length = 1.
x_max = 2*pi
x_min = 0
step = 0.01

#Parameters second function
amplitude2 = .5
frequency2 = 1.
phase2 = 0.
wave_length2 = 2.
x_max2 = 2*pi
x_min2 = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitude, frequency, phase, time,wave_length) for time in x]
y2 = [sine_wave(amplitude2, frequency2, phase2, time,wave_length2) for time in x]
#Composition of first and second function
y = [y[i]+y2[i] for i in range(len(y))]
plt.style.use('dark_background')
plt.plot(x,y)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.show()
```

## Importancia de las señales senoidales y compuestas
Las señales senoidales son producidas por varios fenómenos naturales, que se utilizan para algún procesamiento. Muchas de estas señales son compuestas.
### Uso código (Modificación de parámetros)
Cambiese las variables después de los comentarios "first function" y "second function" para notar otra composición de onda


```python
import scipy.signal as sc

def sine_wave(A: float, f: float, phi: float, t: float,wl: float) -> float:
    return A*sin(2*pi*wl*f*t + phi)

def cos_wave(A: float, f: float, phi: float, t: float,wl: float) -> float:
    return A*cos(2*pi*wl*f*t + phi)

#Parameters first function
amplitude = 1.
frequency = 1.
phase = 0.
wave_length = 1.
x_max = 2*pi
x_min = 0
step = 0.01

#Parameters second function
amplitude2 = .5
frequency2 = 1.
phase2 = 0.
wave_length2 = 2.
x_max2 = 2*pi
x_min2 = 0
step = 0.01
m = 1

x = np.arange(x_min, x_max, step)
y = [sine_wave(amplitude, frequency, phase, time,wave_length) for time in x]
y2 = [cos_wave(amplitude2, frequency2, phase2, time,wave_length2) for time in x]
#Composition of first and second function
y = [(1+m*(y[i]/amplitude))*y2[i] for i in range(len(y))]
plt.style.use('dark_background')
plt.plot(x,y)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo (s)')
plt.grid(True)
plt.show()
```

## Ancho de banda

El ancho de banda de una señal compuesta es la diferencia entre la frecuencia más alta y más baja contenidas en la señal. Es decir, para calcular el ancho de banda es necesario sustraer la frecuencia más baja de la frecuencia más alta del rango.

## Transformación de Pourier
La transformación de Fourier es un método analítico principalmente orientado a funciones integrables que pertenecen a la familia de las transformadas integrales. Este método consiste en una redefinición de funciones f (t) en términos de Cos (t) y Sen (t).

Las identidades trigonométricas de dichas funciones, en conjunto con sus características de derivación y antiderivación, sirven para definir la transformada de Fourier a través de la siguiente función compleja:

F [f(t)](z)=-F(t)e-itdt
