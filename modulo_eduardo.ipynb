```python
from math import sin, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

%matplotlib inline

#Graficar una seal senoidal
def sinu(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amp = 1
freq = 1
phase = 1
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sinu(amp, freq, phase, time) for time in x]


plt.style.use('dark_background')
plt.plot(x,y)
plt.title("seal senoidal")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()
```

```python

#Graficar el concepto periodo de una seal senoidal
def sinu(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

def ind_max_fun(x):
    i = 0
    maxs = []
    max = 0
    for i in range(len(x)):
        if(x[i] >= max):
            maxs.append(i)
            max = x[i]
    return maxs


amp = 1
freq = 2
phase = 1
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sinu(amp, freq, phase, time) for time in x]

local_max = ind_max_fun(y)

plt.style.use('dark_background')
plt.plot(x,y)
plt.plot(x[local_max[len(local_max)-1]],y[local_max[len(local_max)-1]],'ro-')
plt.plot(x[local_max[len(local_max)-2]],y[local_max[len(local_max)-2]],'ro-')
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()
```

```python
#Graficar el concepto de frecuencia de una seal senoidal

def sinu(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amp = 1
freq = 3
phase = 1
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sinu(amp, freq, phase, time) for time in x]

plt.annotate('Frecuencia = ', xy=(2.2, 1), xytext=(3, 1.5),arrowprops=dict(arrowstyle='->'),)
plt.style.use('dark_background')
plt.plot(x,y)
plt.text(4.3, 1.5 , freq)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()
```

```python

#Graficar el concepto de amplitud de una seal senoidal
def sinu(A: float, f: float, phi: float, t: float) -> float:
    return A*sin(2*pi*f*t + phi)

amp = 2
freq = 1
phase = 1
x_max = 2*pi
x_min = 0
step = 0.01

x = np.arange(x_min, x_max, step)
y = [sinu(amp, freq, phase, time) for time in x]

plt.annotate('Amplitud = ', xy=(0, amp), xytext=(0, amp+0.5),arrowprops=dict(arrowstyle='->'),)
plt.style.use('dark_background')
plt.plot(x,y)
plt.text(1.2, amp+0.5 , amp)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()
```
